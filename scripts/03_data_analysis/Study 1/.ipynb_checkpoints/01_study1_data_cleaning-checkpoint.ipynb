```python
df = pd.read_csv("./01_raw_data_anonymized/study1_prolific_anonymized.csv")
```

```python
# Identify pairs of columns to combine based on the mapping you provided
pairs = [
    ("C_Message Rank_" + str(i), "T_Message Ranking_" + str(i)) for i in range(1, 31)
] + [
    ("C_Dis_AI" + str(i), "T_Dis_AI" + str(i) + "_1") for i in range(1, 16)
] + [
    ("C_Dis_Hum" + str(i), "T_Dis_Hum" + str(i) + "_1") for i in range(1, 16)
] + [
    ("C_Con_AI" + str(i), "T_Con_AI" + str(i) + "_1") for i in range(1, 16)
] + [
    ("C_Con_Hum" + str(i), "T_Con_Hum" + str(i) + "_1") for i in range(1, 16)
] + [
    ("C_Unp_AI" + str(i), "T_Unp_AI" + str(i) + "_1") for i in range(1, 16)
] + [
    ("C_Unp_Hum" + str(i), "T_Unp_Hum" + str(i) + "_1") for i in range(1, 16)
] + [
    ("C_Lapp_AI" + str(i), "T_Lapp_AI" + str(i) + "_1") for i in range(1, 16)
] + [
    ("C_Lapp_Hum" + str(i), "T_Lapp_Hum" + str(i) + "_1") for i in range(1, 16)
]

# Create new columns without modifying the original DataFrame in a loop
new_cols = {}
for c_col, t_col in pairs:
    combined_col_name = c_col.replace('C_', '')
    new_cols[combined_col_name] = df[c_col].fillna(df[t_col])

# Concatenate the new columns to the original DataFrame
df = pd.concat([df, pd.DataFrame(new_cols)], axis=1)
```

```python
df2 = df
cols2 = df2.columns.tolist()
cols2.insert(0, cols2.pop(cols2.index('Experimental_Group')))
df2 = df2[cols2]
df2['Experimental_Group'] = df2['Experimental_Group'].map({0: 'Not Disclosed', 1: 'Disclosed'})
df2["Duration (in seconds)"] = pd.to_numeric(df2["Duration (in seconds)"])
df2 = df2[(df2["Duration (in seconds)"] > 300) & (df2["Finished"] == 1)]
# Find the index positions of 'Finished' and 'Age'
Finished_idx = cols2.index('Finished')
age_idx = cols2.index('Age')
# Reorder the columns
new_cols_order = cols2[:Finished_idx + 1] + cols2[age_idx:] + cols2[Finished_idx + 1:age_idx]
# Reassign the columns to the DataFrame based on the new order
df2 = df2[new_cols_order]
df2.head()
```

```python
df2.to_csv('./02_cleaned_data_file/study1_prolific_cleaned.csv', index=False)
```

```python
df_sona = pd.read_csv("./01_raw_data_anonymized/study1_sona_anonymized.csv")
```

```python
cols_sona = df_sona.columns.tolist()

# Extract unique numbers from C_Message Rank_ columns
message_rank_numbers = sorted([int(col.split('_')[-1]) for col in cols_sona if col.startswith('C_Message Rank_')])

# Generate pairs based on actual columns in dataframe for C_Message Rank_ and T_Message Ranking_
message_pairs = [("C_Message Rank_" + str(i), "T_Message Ranking_" + str(i)) for i in message_rank_numbers]

# Other pairs remain the same
other_pairs = [
    ("C_Dis_AI_" + str(i), "T_Dis_AI_" + str(i)) for i in range(1, 16)
] + [
    ("C_Dis_hum_" + str(i), "T_Dis_hum_" + str(i)) for i in range(1, 16)
] + [
    ("C_Con_AI_" + str(i), "T_Con_AI_" + str(i)) for i in range(1, 16)
] + [
    ("C_Con_hum_" + str(i), "T_Con_hum_" + str(i)) for i in range(1, 16)
] + [
    ("C_Unp_AI_" + str(i), "T_Unp_AI_" + str(i)) for i in range(1, 16)
] + [
    ("C_Unp_hum_" + str(i), "T_Unp_hum_" + str(i)) for i in range(1, 16)
] + [
    ("C_Lapp_AI_" + str(i), "T_Lapp_AI_" + str(i)) for i in range(1, 16)
] + [
    ("C_Lapp_hum_" + str(i), "T_Lapp_hum_" + str(i)) for i in range(1, 16)
]

# Combine pairs
pairs2 = message_pairs + other_pairs

# Create new columns without modifying the original DataFrame in a loop
new_cols_sona = {}
for c_col, t_col in pairs2:
    combined_col_name = c_col.replace('C_', '')
    new_cols_sona[combined_col_name] = df_sona[c_col].fillna(df_sona[t_col])

# Concatenate the new columns to the original DataFrame
df_sona = pd.concat([df_sona, pd.DataFrame(new_cols_sona)], axis=1)
```

```python
df_sona2 = df_sona
df_sona2['Experimental_Group'] = np.where(df_sona2['T_Message Ranking_1'].isna(), 'Not Disclosed', np.where(df_sona2['C_Message Rank_1'].isna(), 'Disclosed', np.nan))
cols_sona2 = df_sona2.columns.tolist()
cols_sona2.insert(0, cols_sona2.pop(cols_sona2.index('Experimental_Group')))
df_sona2 = df_sona2[cols_sona2]
df_sona2["Duration (in seconds)"] = pd.to_numeric(df_sona2["Duration (in seconds)"])
df_sona2 = df_sona2[(df_sona2["Duration (in seconds)"] > 300) & (df_sona2["Finished"] == 1)]
# Find the index positions of 'Finished' and 'Age'
sona_Finished_idx = cols_sona2.index('Finished')
sona_age_idx = cols_sona2.index('Age')
# Reorder the columns
new_cols_order2 = cols_sona2[:sona_Finished_idx + 1] + cols_sona2[sona_age_idx:] + cols_sona2[sona_Finished_idx + 1:sona_age_idx]
# Reassign the columns to the DataFrame based on the new order
df_sona2 = df_sona2[new_cols_order2]
df_sona2.head()
```

```python
df_sona2 = df_sona2.rename(columns=lambda x: x.replace('_hum', '_Hum') if '_hum' in x else x)
```

```python
df_sona2.to_csv('./02_cleaned_data_file/study1_sona_cleaned.csv', index=False)
```
