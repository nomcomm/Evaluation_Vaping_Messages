# Step 2: Clean Combined Data File

After creating the combined excel file, the next task is to clean the data file. 

```python
#Import the Python packages needed for the cleaning.
import pandas as pd
import numpy as np
```

```python
#Import the combined tweets file and 
vaping_tweets = pd.read_csv('/Users/suelim/anaconda3/Message_Engine3/data/NCA 2023/00_Scraped_messages/scraped_tweets_combined.csv')

#Put the tweets column into the tweets column and then count how many tweets there are.
tweets = vaping_tweets['rawContent'].values
print(len(tweets))

#List the columns that are in the scraped twitter file.
vaping_tweets.columns
```

```python
#Sort tweets by "retweetCount" column. This is because we want to
#select messages that are most retweeted (aka. should be effective messages)

sorted_tweets = vaping_tweets.sort_values(by="retweetCount", ascending=False)

sorted_tweets
```

```python
#Delete tweets that have not been retweeted. Make sure to use sorted_tweets dataframes

retweeted_tweets = sorted_tweets[sorted_tweets.retweetCount > 0].values

result = []
for item in tweets:
    if item in retweeted_tweets:
        result.append(item)
        
print(len(result))
```

```python
#Delete tweets in English language

en_tweets = sorted_tweets[sorted_tweets.lang == 'en'].values

result2 = []
for item in result:
    if item in en_tweets:
        result2.append(item)
        
print(len(result2))
```

```python
#Delete duplicate items

seen = set()
result3 = []
for item in result2:
    if item not in seen:
        seen.add(item)
        result3.append(item)#.encode(encoding='UTF-8',errors='strict'))
print(len(result3))
```

```python
#Take out the symbols and keep only the text
import re
result4 = []
for currenttext in result3:
    text = re.sub(r"http\S+", "", currenttext)
    text = re.sub(r"@\S+", "", text)
    result4.append(text)
print(len(result4))
```

```python
#Remove the white space that are in the tweets

def remove_extra_whitespace_tabs(text):
    #pattern = r'^\s+$|\s+$'
    pattern = r'^\s*|\s\s*'
    return re.sub(pattern, ' ', text).strip()

seen2 = set()
result5 = []
for item in result4:
    if item not in seen2:
        seen2.add(item)
        clean_str = ''.join([c for c in item if ord(c) < 128])
        cs2 = remove_extra_whitespace_tabs(clean_str)
        result5.append(cs2)#.encode())
print(len(result5))
```

```python
#Delete tweets with advertisements and are too short

result6 = []
for i in range(len(result5)):
    if 'shop' in result5[i]:
        print('shop!')
    elif 'buy' in result5[i]:
        print('buy!')  
    elif (len(result5[i]) < 30):
        print(result5[i])  
    else:
        result6.append(result5[i])
print(len(result6))      
```

```python
#Finally, we save the tweets that are left into a csv file and put in data folder
dedup_df = pd.DataFrame(result6, columns = ['tweet'])
dedup_df.to_csv('/Users/suelim/anaconda3/Message_Engine3/data/NCA 2023/00_Scraped_messages/vaping_tweets_cleaned.csv', index = False)
```
