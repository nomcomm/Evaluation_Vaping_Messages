```python
def clean_text(text):
   #Make text lowercase   
   text = text.lower()
   #remove text in square brackets
   text = re.sub(r'\[.*?\]', '', text)
   #remove punctuation   
   text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text) 
   #remove words containing numbers
   text = re.sub(r'\w*\d\w*', '', text)
   return text
```

```python
#without prompt
def get_top_n_words(corpus, n=None):
    vec = CountVectorizer(stop_words='english').fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]


common_words = get_top_n_words(all_humanmsg_df.tweets, 10)
human_unigram = pd.DataFrame(common_words, columns = ['unigram' , 'count'])

common_words = get_top_n_words(all_bloomgenmsg_np_df.tweets, 10)
aigen_unigram = pd.DataFrame(common_words, columns = ['unigram' , 'count'])

print(human_unigram)
print('---')
print(aigen_unigram)
```

```python
def get_top_n_trigram(corpus, n=None):
    vec = CountVectorizer(ngram_range=(2,2),stop_words='english')
    vec.fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in      vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]

common_words = get_top_n_trigram(all_humanmsg_df.tweets, 10)
human_trigram = pd.DataFrame(common_words, columns = ['bigram' , 'count'])
print(human_trigram)


common_words = get_top_n_trigram(all_bloomgenmsg_np_df.tweets, 10)
aigen_trigram = pd.DataFrame(common_words, columns = ['bigram' , 'count'])
print(aigen_trigram)

```

### Topic Modeling

```python
from sklearn.decomposition import LatentDirichletAllocation

human_vectorizer =  CountVectorizer(analyzer='word',       
                    min_df=3,# minimum required occurences of a word 
                    stop_words='english',# remove stop words
                    lowercase=True,# convert all words to lowercase
                    token_pattern='[a-zA-Z0-9]{3,}',# num chars > 3
                    max_features=5000,# max number of unique words
                     )

human_data_matrix = human_vectorizer.fit_transform(all_humanmsg_df.tweet)

human_lda_model   = LatentDirichletAllocation(
                    n_components=10, # Number of topics
                    learning_method='online',
                    random_state=20,       
                    n_jobs = -1  # Use all available CPUs
                    )

human_lda_output = human_lda_model.fit_transform(human_data_matrix)




aigen_vectorizer  = CountVectorizer(analyzer='word',       
                    min_df=3,# minimum required occurences of a word 
                    stop_words='english',# remove stop words
                    lowercase=True,# convert all words to lowercase
                    token_pattern='[a-zA-Z0-9]{3,}',# num chars > 3
                    max_features=5000,# max number of unique words
     )

aigen_data_matrix = aigen_vectorizer.fit_transform(all_aigenmsg_df.tweet)

aigen_lda_model =   LatentDirichletAllocation(
                    n_components=10, # Number of topics
                    learning_method='online',
                    random_state=20,       
                    n_jobs = -1  # Use all available CPUs
                    )

aigen_lda_output = aigen_lda_model.fit_transform(aigen_data_matrix)
```

### similarity

```python
result = pd.concat([selected_aigenmsg_df, selected_humanmsg_df])
result.head()
#len(result)
```

```python
import spacy
import wmd

nlp = spacy.load('en_core_web_md')# create_pipeline=wmd.WMD.create_spacy_pipeline)
```

```python
sims  = np.zeros((60,60))
wmds  = np.zeros((60,60))

for current_i in range(60):
    print(current_i)
    print('---')
    #sys.stdout.write('Current AI Msg.: ') 
    #sys.stdout.write(str(current_aigen) + ': ')     
    doc1 = nlp(result.tweet.iloc[current_i])
    #print(result.tweet.iloc[current_i])
       
    for current_j in range(60):
        print(current_i)
        print(current_j)
        doc2 = nlp(result.tweet.iloc[current_j])   
        #sys.stdout.write('Current Human Msg.: ') 
        #sys.stdout.write(str(current_human) + ': ')     
        #print(selected_humanmsg_df.tweet[current_human])
        sim_value = doc1.similarity(doc2)
        sims[current_i, current_j] = sim_value
        
    
        wmd_instance = wmd.WMD.SpacySimilarityHook(nlp)
        wmds[current_i, current_j]  = wmd_instance.compute_similarity(doc1, doc2)
        #sys.stdout.write('Sim Value: ') 
        #print(sim_value)
    print('---')
```

```python
plt.figure(figsize = (8,8))
plt.imshow(sims, cmap = "seismic", vmin = 0.4, vmax = 1.1)
plt.colorbar(shrink=0.55);

plt.figure(figsize = (8,8))
plt.imshow(wmds, cmap = "seismic", vmin = 0, vmax = 10)
plt.colorbar(shrink=0.55);
```

```python
from nilearn import connectome
sims_vec = nilearn.connectome.sym_matrix_to_vec(sims, discard_diagonal=True)
wmds_vec = nilearn.connectome.sym_matrix_to_vec(wmds, discard_diagonal=True)

sims_vec.shape
```

```python
plt.scatter(sims_vec, wmds_vec)
np.corrcoef(sims_vec, wmds_vec)
```
