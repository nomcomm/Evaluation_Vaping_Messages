```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('./data/study1_Ralf.csv')
df.head()

df1 = pd.DataFrame()
df1 = df.groupby('Participant').agg({'Effects_Perception': 'mean',
                                     'recruitment_platform': 'first',
                                     'Experimental_Group': 'first'})
df1['question'] = 'EP'                   


sns.violinplot(data=df1, x="question", y="Effects_Perception", hue="recruitment_platform",
               split=True, inner="quart", linewidth=1, palette={"Prolific": "salmon", "SONA": "dodgerblue"})


df1.head()
```

```python
df = pd.read_csv('./data/study2_Ralf.csv')
df.head()
```

```python


df2 = pd.DataFrame()
df2 = df.groupby('Participant').agg({'Effects_Perception': 'mean',
                                     'AIper_neg': 'mean',
                                     'recruitment_platform': 'first',
                                     'Experimental_Group': 'first'})
df2['question'] = 'EP' 
df2['question2'] = 'AIper_neg'


sns.violinplot(data=df2, x="question", y="Effects_Perception", hue="recruitment_platform",
               split=True, inner="quart", linewidth=1, palette={"Prolific": "salmon", "SONA": "dodgerblue"})


df2.head()
```
